Based on EDA we need to answer some questions:
  * Which gender is more likely to default (male, female)?
  * Which class get paid more money?
  * What is the relation between applicant age and defaulting?
  * What is the relation between years of experience and defaulting?
  * What is The relation between occupation and defaulting?
  * Does having a car, house or kids affect the applicant ability to repay on time?
  * Does living in a specific area or city affect the applicant ability to repay on time?
  * Does having multiple previous Credit Bureau affect the ability to repay on time?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
# loading datasets
train_data = pd.read_csv("dataset/application_train.csv")
```

```python
target_count = pd.DataFrame(train_data.TARGET.values.tolist()).stack().value_counts()
labels = [0,1]
plt.pie(target_count, labels=labels, autopct='%0.f%%', startangle=70)
pai = plt.gcf()
plt.axis('equal')
plt.tight_layout()
```

Dataset is highly unbalanced, we could work on resampling (Oversampling | Undersampling)

```python
fig, ax = plt.subplots(figsize=(7,7))
ax = sns.countplot(x='CODE_GENDER', data=train_data)
for p in ax.patches:
        ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.0, p.get_height()+50))
        ax.annotate('%{:.1f}'.format((p.get_height()/len(train_data))*100), (p.get_x()+0.5, p.get_height()+50))
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax = sns.countplot(x='TARGET', hue='CODE_GENDER', data=train_data)
for p in ax.patches:
        ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.0, p.get_height()+50))
        ax.annotate('%{:.1f}'.format((p.get_height()/len(train_data))*100), (p.get_x()+0.5, p.get_height()+50))
```

```python
train_data['DAYS_BIRTH'] = abs(train_data['DAYS_BIRTH'])
plt.hist(train_data['DAYS_BIRTH'] / 365, edgecolor = 'y', bins = 27)
plt.title('Usr age'); plt.xlabel('Age (years)'); plt.ylabel('Count');
```

```python
plt.figure(figsize = (13, 13))
# Repaid on time
sns.kdeplot(train_data.loc[train_data['TARGET'] == 0, 'DAYS_BIRTH'] / 365, label = 'target == 0')
# NOT repaid on time
sns.kdeplot(train_data.loc[train_data['TARGET'] == 1, 'DAYS_BIRTH'] / 365, label = 'target == 1')
plt.xlabel('Age (years)'); plt.ylabel('Density')
plt.title('Ages Distribution')
```
